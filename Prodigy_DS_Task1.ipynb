Top 10 countries from total

```python
# Filter data for total population
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Sort data based on the total population for 2020
total_population_sorted = total_population_data.sort_values(by="2020", ascending=False)

# Get the top ten countries with the highest total population for 2020
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Code']])
```

```python
xxxxxxxxxxxxxxxxxxxxxplt.figure(figsize=(13, 7))
plt.subplot(2,2,1)
sns.barplot(x="2020", y="Country Code", data=total_top_ten_countries, palette="rocket")
plt.title("Top Ten Countries of Total Population (2020)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(13, 7))
plt.subplot(2,2,2)
sns.barplot(x="2015", y="Country Code", data=total_top_ten_countries, palette="rocket")
plt.title("Top Ten Countries of Total Population (2015)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

top 10 countries with highest male population

```python
#filter data for male population
male_population_data = df[df['Series Code']=='SP.POP.TOTL.MA.IN']

#sort data based on the male population for 2020
male_population_sorted =male_population_data.sort_values(by="2020",ascending=False)

#Get the top 10 countries with the highest male population for 2020
male_top_ten_countries = male_population_sorted .head(10)
print("Top 10 countries of male population")
print(male_top_ten_countries[['Country Code']])
```

top 10 countries with highest female population

```python
#filter data for female population
female_population_data = df[df['Series Code']=='SP.POP.TOTL.FE.IN']

#sort data based on the female population for 2020
female_population_sorted =female_population_data.sort_values(by="2020",ascending=False)

#Get the top 10 countries with the highest female population for 2020
female_top_ten_countries = female_population_sorted .head(10)
print("Top 10 countries of female population")
print(female_top_ten_countries[['Country Code']])
```

Bar plot for highest male and female population in 2020

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2020", y="Country Code", data=male_top_ten_countries, palette="rocket")
plt.title("Top 10 countries of male population(2020)",fontsize=10)
plt.xlabel("Male Population",fontsize=10)
plt.ylabel("Country",fontsize=10)

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2015", y="Country Code", data=female_top_ten_countries, palette="rocket")
plt.title("Top 10 countries of female population(2015)",fontsize=10)
plt.xlabel("Female Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

Stacked bar plot Top 10 countries with male and female population

```python
#merge male and female population data on 'Country Code'
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))
```

```python
merge_data["Total population"] = merge_data["2022_male"] + merge_data ["2022_female"]
```

```python
merge_data.head()
```

```python
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

Stacked graph for top 10 countries total population of male and female

```python
top_10_countries = sorted_data.head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2020_female",data=top_10_countries,color="pink",label="female Population")
sns.barplot(x="Country Code",y="2020_male",data=top_10_countries,bottom=top_10_countries["2020_female"],color="skyblue",label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

Stacked graph for Bottom 10 countries total population of male and female

```python
bottom_10_countries = sorted_data.tail(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Country Code", y="2020_female", data=bottom_10_countries, color="pink", label="Female Population")
sns.barplot(x="Country Code", y="2020_male", data=bottom_10_countries, bottom=bottom_10_countries["2020_female"], color="skyblue", label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
